```python
# %%
from sklearn import linear_model
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats

# %%
resp = make_regression(30, 1, noise=10)
x, y = resp

# %%
type(x), y.shape

# %%
w0, w1, *rest =  stats.linregress([i[0] for i in x], y)
print(w0, w1)

# %%
def func(x):
    return w0 * x + w1

# %%
plt.scatter(x, y)
plt.plot(x, [ func(i[0]) for i in x])
```
