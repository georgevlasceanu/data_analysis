```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
```

```python
x = [10, 20, 30, 40, 50]
y = [5, 10, 15, 20, 25]
plt.scatter(x, y, c="red")

def f(x):
    return x / 2
model = list(map(f, x))

plt.plot(x, model)
```

```python
x = [10, 20, 30, 40, 50]
y = [5, 10, 15, 20, 25]
w0, w1, *rest = stats.linregress(x, y)
w0, w1
```

```python
def func(x):
    return x * w0 + w1
```

```python
plt.scatter(x, y)
plt.plot(x, [func(i) for i in x])
```
