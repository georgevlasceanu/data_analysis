```python
apartamente = [(45, 60), (70, 115), (120, 170)]

x = [i[0] for i in apartamente]
y = [i[1] for i in apartamente]

def func1(x):
    return 1.5 * x

model1 = [func1(i) for i in x]

plt.plot(x, model1)

def func2(x):
    return 1.6 * x
model2 = [func2(i) for i in x]

plt.plot(x, model2, c="green")


def func3(x):
    return 1.4 * x
model3 = [func3(i) for i in x]

plt.plot(x, model3, c="fuchsia")

plt.scatter(x, y)
plt.xlabel("Metri patrati")
plt.ylabel("Valoarea in Euro")
plt.show()
```

```python
from scipy import stats
```

```python
apartamente = [(45, 60), (70, 115), (120, 170)]

x = [i[0] for i in apartamente]
y = [i[1] for i in apartamente]

result = stats.linregress(x, y)
print(result)
```

```python
slope, intercept = result
```

```python
def func_regresie(x):
    return x * slope + intercept
```
