```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
```

```python
apartamente = [(45, 60), (70, 115), (120, 170), (82.6, 170), (44, 80), (74.3, 110), (81, 198), (117, 180), (50, 88), (54, 114)]

x = [i[0] for i in apartamente]
y = [i[1] for i in apartamente]
w0, w1, *rest = stats.linregress(x, y)
w0, w1
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x, y)
model = list(map(lambda x: w0 * x + w1, x))

model = [w0 * i + w1 for i in x]

plt.plot(x, model)
```
