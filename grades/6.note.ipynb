```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("note.csv")
```

```python
ore = df["Ore_Studiate"]
ore = df.iloc[:,0:1]
note = df["Note"]
```

```python
model = LinearRegression()
```

```python
model.fit(ore, note)
```

```python
note_prezise = model.predict(ore)
```

```python
plt.scatter(ore, note)
plt.scatter(ore, note_prezise)
plt.plot(ore, note)
plt.plot(ore, note_prezise)
```
